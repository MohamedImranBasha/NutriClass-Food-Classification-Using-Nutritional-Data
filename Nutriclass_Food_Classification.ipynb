```python
import numpy as np

def cap_outliers_iqr(series):
    q1 = np.percentile(series, 25)
    q3 = np.percentile(series, 75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series.clip(lower=lower, upper=upper)

important_cols = ['Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber',
                  'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content', 'Serving_Size']

for col in important_cols:
   df.loc[:, col] = cap_outliers_iqr(df[col])
```

```python
df[important_cols].describe()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Select the data from the specified columns
df_scaled = scaler.fit_transform(df[important_cols])

# Initialize PCA with a desired number of components
pca = PCA(n_components=2) # Retain 2 principal components

# Fit PCA to the scaled data and transform it
df_pca = pca.fit_transform(df_scaled)

# Explained variance ratio of each principal component
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Cumulative explained variance ratio
print("Cumulative explained variance ratio:", np.sum(pca.explained_variance_ratio_))

# The transformed data with reduced dimensions
print("Transformed data (df_pca):\n", df_pca)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y=df['Food_Name']
y_encoded = le.fit_transform(y)
print(df['Food_Name'].value_counts(normalize=True))
```

```python
##MODEL SELECTION AND TRAINING
##1. logistic regression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score


X_train, X_test, y_train, y_test = train_test_split(df_pca, y_encoded, test_size=0.3, random_state=42)

clf = LogisticRegression(random_state=42, max_iter=1000)
clf.fit(X_train, y_train)

y_pred_lr = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_lr, average='weighted'))
print("f1:", f1_score(y_test, y_pred_lr, average='weighted'))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns

cm = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
##LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Use original scaled features
X = df[important_cols] # Select the data from the important columns
y = y_encoded  # Use the encoded target variable


lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X, y)

feature_names = important_cols

# For multiclass, clf.coef_ shape is (n_classes, n_features)

for i, class_label in enumerate(lr.classes_):
        plt.figure(figsize=(10, 6))
        plt.barh(feature_names, lr.coef_[i])
        plt.title(f"Feature Weights - Class: {le.inverse_transform([class_label])[0]}") # Use inverse_transform to get original class name
        plt.xlabel("Weight")
        plt.ylabel("Features")
        plt.tight_layout()
        plt.show()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import matplotlib.pyplot as plt

model = OneVsRestClassifier(DecisionTreeClassifier(random_state=42))
model.fit(X, y)

# Loop over each binary classifier
for i, class_label in enumerate(model.classes_):
    importances = model.estimators_[i].feature_importances_
    sorted_idx = np.argsort(importances)

    plt.figure(figsize=(10, 6))
    plt.barh(X.columns[sorted_idx], importances[sorted_idx])
    plt.title(f"Feature Importance for class: {class_label}")
    plt.xlabel("Importance")
    plt.ylabel("Features")
    plt.tight_layout()
    plt.show()
```
